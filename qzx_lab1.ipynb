```python
import pandas as pd
import numpy as np
titanic_survival = pd.read_csv("titanic_train.csv")
titanic_survival.head()
```

```python
age = titanic_survival["Age"]
print(age.loc[0:10])
age_is_null = pd.isnull(age)
print (age_is_null.loc[0:10])
age_null_true = age[age_is_null]
print (age_null_true)
print (age[age_is_null])
age_null_count = len(age_null_true)
print(age_null_count)
```

```python
print(age.mean())
```

```python
passenger_fare = titanic_survival.pivot_table(index="Pclass", values="Fare")
print(passenger_fare)
```

```python
port_stats = titanic_survival.pivot_table(index="Embarked", values=["Survived"], aggfunc=np.mean)
print(port_stats)
```

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
column_null_count = titanic_survival.apply(not_null_count)
print (column_null_count)
```

```python
def is_minor(row):
    if row["Age"] < 18:
        return True
    else:
        return False
minors = titanic_survival.apply(is_minor, axis=1)
def generate_age_label(row):
    age = row["Age"]
    if pd.isnull(age):
        return "unknown"
    elif age < 18:
        return "minor"
    else:
        return "adult"
age_labels = titanic_survival.apply(generate_age_label, axis=1)
print (age_labels)
```
